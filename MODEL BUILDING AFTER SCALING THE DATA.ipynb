```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#Scaling with labelencoding

```python
data=pd.read_csv('Insurance_Marketing-Customer-Value-Analysis.csv')
```

```python
data.head()
```

```python
data.drop(['Customer','Effective To Date'],axis=1,inplace=True)
```

```python
data.columns
```

```python
for x in data.columns:
    if len(x.split(' '))>1:
        a=x.split(' ')
        b=str.join('_',a)
        data.rename(columns={x:b},inplace=True)
```

```python
data.columns
```

```python
data.info()
```

```python
data_object=data.select_dtypes('object')
```

```python
data_object
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data_object=data_object.apply(le.fit_transform)
```

```python
data_object.head()
```

```python
data_num=data.select_dtypes(exclude='object')
```

```python
data_num.head()
```

```python
col=data_num.columns
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss.fit(data_num)
```

```python
data_num=ss.transform(data_num)
```

```python
data_num=pd.DataFrame(data_num)
data_num.columns=col
```

```python
data_num.head()
```

```python
data_num.describe()
```

```python
data_scale=pd.concat([data_object,data_num],axis=1)
```

```python
data_scale
```

```python
X=data_scale.drop('Customer_Lifetime_Value',axis=1)
y=data_scale['Customer_Lifetime_Value']
```

```python
import warnings 
warnings.filterwarnings('ignore')
import statsmodels.api as sm

X_constant = sm.add_constant(X)
lin_reg = sm.OLS(y,X_constant).fit()
lin_reg.summary()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
def model_fit(name,algo,X_train,X_test,y_train,y_test):
    print(name)
    algo.fit(X_train,y_train)
    print('train score {}'.format(algo.score(X_train,y_train)))
    print('test score {}'.format(algo.score(X_test,y_test)))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import AdaBoostRegressor
DT = DecisionTreeRegressor(criterion = 'entropy',random_state= 0)
DT_reg=DecisionTreeRegressor(max_depth=5,criterion = 'entropy',random_state= 0)
RF=RandomForestRegressor(n_estimators=10,random_state=0)
LR=LogisticRegression()
knn=KNeighborsRegressor()
KNN_bag=BaggingRegressor(base_estimator=knn,n_estimators=10,random_state=0)
LR_bag=BaggingRegressor(base_estimator=LR,n_estimators=17,random_state=0)
DT_bag=BaggingRegressor(n_estimators=10,random_state=0)
LR_boost=AdaBoostRegressor(base_estimator=LR,n_estimators=50,random_state=0)
DT_boost=AdaBoostRegressor(n_estimators=50,random_state=0)
RF_boost=AdaBoostRegressor(base_estimator=RF,n_estimators=50,random_state=0)
model=[]
model.append(('Logistic Regression',LR))
model.append(('Decision Tree',DT))   
model.append(('Decision Tree Regularized',DT_reg)) 
model.append(('Random Forest',RF))
model.append(('KNeighbors',knn))
model.append(('KNN_BAG',KNN_bag))
model.append(('LR_BAG',LR_bag))
model.append(('DT_BAG',DT_bag))
model.append(('LR_BOOST',LR_boost))
model.append(('DT_BOOST',DT_boost))
model.append(('RF_BOOST',RF_boost))
```

```python
for x,y in model:
    model_fit(x,y,X_train,X_test,y_train,y_test)
```
